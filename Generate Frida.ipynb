```python
%matplotlib inline
from PIL import Image
import matplotlib.pyplot as plt
import os
import random
```

We want to generate Frida Khalo images using a GAN. First, we need to examine our data and add a border to fit to the largest size.

```python
paintings = os.listdir("paintings")
paintings = [Image.open('paintings/'+p) for p in paintings if p != '.DS_Store']
print("Number of paintings",len(paintings))
```

```python
def find_max_w_h(paintings):
    maxw, maxh = 0,0
    for painting in paintings:
        w,h = painting.size
        if w>maxw:
            maxw=w
        if h>maxh:
            maxh=h
    return maxw, maxh

def find_min_w_h(paintings):
    minw, minh = 10000, 10000
    for painting in paintings:
        w,h = painting.size
        if w<minw:
            minw=w
        if h<minh:
            minh=h
    return minw, minh
    
```

They all have the same x axis, but the height varies. 

```python
print("Max values are:", find_max_w_h(paintings))
print("Min values are:", find_min_w_h(paintings))
```

```python
def display_random_images(image_list, n, ncol=3, figsize=(10,10), seed=10):
    image_list = random.sample(image_list,n)
    fig = plt.figure(figsize=figsize)
    column = 0
    for i in range(len(image_list)):
        column += 1
        #  check for end of column and create a new figure
        if column == ncol+1:
            fig = plt.figure(figsize=figsize)
            column = 1
        fig.add_subplot(1, ncol, column)
        plt.imshow(image_list[i])
        plt.axis('off')
    plt.show()
```

```python
display_random_images(paintings, 3)
```

There are various ways we can chose to handle images of different sizes:
* Resize the images to be the same size
* Add black edges to make all images the same size
* Crop images from the center

```python
def add_black_edges_images(image_list,neww, newh):
    new_images = []
    for i in image_list:
        new_images.extend([add_black_edges(i, neww, newh)])
    return new_images

def add_black_edges(image, neww, newh):
    old_size = image.size
    new_size = (neww, newh)

    if new_size > old_size:
        # Set number of pixels to expand to the left, top, right,
        # and bottom, making sure to account for even or odd numbers
        if old_size[0] % 2 == 0:
            add_left = add_right = (new_size[0] - old_size[0]) // 2
        else:
            add_left = (new_size[0] - old_size[0]) // 2
            add_right = ((new_size[0] - old_size[0]) // 2) + 1

        if old_size[1] % 2 == 0:
            add_top = add_bottom = (new_size[1] - old_size[1]) // 2
        else:
            add_top = (new_size[1] - old_size[1]) // 2
            add_bottom = ((new_size[1] - old_size[1]) // 2) + 1

        left = 0 - add_left
        top = 0 - add_top
        right = old_size[0] + add_right
        bottom = old_size[1] + add_bottom

        # By default, the added pixels are black
        image = image.crop((left, top, right, bottom))
        return image
```

```python
transformed_abe = add_black_edges_images(paintings, 300, 514)
display_random_images(transformed_abe, n=3)
```

```python
def crop_image_center(image, newh, neww):
    old_size = image.size
    new_size = (neww, newh)
    
    subtract_top = subtract_bottom = (old_size[1] - new_size[1]) // 2
    image = image.crop((0, 0 + subtract_top, neww, old_size[1] - subtract_bottom))
    return image

def crop_images_center(image_list, newh, neww):
    new_images = []
    for i in image_list:
        new_images.extend([crop_image_center(i, neww, newh)])
    return new_images
```

```python
transformed_cic = crop_images_center(paintings, 150, 300)
display_random_images(transformed_cic, n=3)
```

```python
def resize_images(image_list, neww, newh):
    new_images = []
    
    for i in image_list:
        new_images.extend([i.resize((neww, newh),Image.LANCZOS)])
    return new_images
```

```python
transformed_r = resize_images(paintings, 300, 520)
display_random_images(transformed_r, n=3)
```
